```python
#Let's create new dataframe and useless columns add those to data frame
#Data frame not useless
df_nu = pd.DataFrame()
```

### 1) Columns which number of unique values is [0 , 1] or greater than 70% length of original data frame 

```python
#Let's add
for i,v in zip(df.nunique().index , df.nunique()):
    if v in [0,1] or v > 1000:
       df_nu[i] = df[i]
```

```python
#Let's look at data frame
df_nu
```

```python
#Let's find columns which we add as useless columns
useless = list(df_nu.columns)
```

## 2) Columns which have imbalance problems

```python
#Let's look at imbalanced features
for i,v in zip(df.nunique().index , df.nunique()):
    if v == 2:
        if df[i].value_counts()[1] < df[i].value_counts()[0]*0.04:
            print(df[i].value_counts())
            if i not in useless:
                useless.append([i])
            print('=============================')
```

```python
#List which contain Useless columns
useless
```

## 3) Columns which have low correllation with churn column

```python
#Let's look at correlations of columns with target columns 
for i in df.select_dtypes('int').columns:
    corr = df[i].corr(df['churn'])
    if abs(corr) < 0.02:
        if i not in useless:
            useless.append(i)
        print(f'{i} & Churn ==> ,Correlation: {corr}')
```

```python
#Let's normalize columns
for i in df.select_dtypes('int'):
    df[i] = (df[i] - df[i].min())/(df[i].max() - df[i].min())
```

```python
#Let's encode categorical columns
from sklearn.preprocessing import LabelEncoder
for i in df.select_dtypes(exclude = ['int','float']).columns:
    df[i] = LabelEncoder().fit_transform(df[i])  
```

```python
#Let's check general number of columns
len(useless)
```

```python
#Let's examine null values
for i in df.isnull().sum():
    if i>1:
        print(i)
```

```python
#Data Frame
df
```

```python
#Let's create data frame containing with useless columns
for i in useless:
    df_nu[i] = df[i]
```

```python
df_target = df['churn']
del df['churn']
```

```python
#Let's import PCA library  
from sklearn.decomposition import PCA
```

```python
#Let's fit df to PCA
pca = PCA(n_components = 1).fit(df_nu)
```

```python
#Let's examine what the optimal number of components is 
n_components = 1
while pca.explained_variance_ratio_.cumsum()[-1] < 0.95:
    pca = PCA(n_components = n_components).fit(df_nu)
    print(f'Components ==> {n_components} : {pca.explained_variance_ratio_.cumsum()}')
    print("======================================================================================")
    n_components += 1
```

```python
#Let's fit df to PCA
pca = PCA(n_components = 1).fit(df_nu)
```

```python
#Let's transform data frame 
df_nu = pd.DataFrame(pca.transform(df_nu))
```

```python
#Let's look at columns
for i in df.columns:
    if i in useless:
        del df[i]
```

```python
#Let's fit df to PCA
pca = PCA(n_components = 1).fit(df)
```

```python
#Let's examine what the optimal number of components is 
n_components = 1
while pca.explained_variance_ratio_.cumsum()[-1] < 0.95:
    pca = PCA(n_components = n_components).fit(df)
    print(f'Components ==> {n_components} : {pca.explained_variance_ratio_.cumsum()}')
    print("======================================================================================")
    n_components += 1
```

```python
#Let's fit df to PCA
pca = PCA(n_components = 8).fit(df)
```

```python
#Let's transform data frame 
df = pd.DataFrame(pca.transform(df))
```

```python
#Let's combine three data frame
df_general = pd.concat([df,df_nu,df_target],axis = 1 ,ignore_index = True)
```

```python
#let's replace name of churn column
df_general.rename({9:'churn'} , axis = 1, inplace = True)
```

```python
#Let's check correlations
for i in df_general:
    if i != 'churn':
        corr = df_general[i].corr(df_general['churn'])
        print(f'{i} & Churn  ==>  Correlation: {corr}')
```

```python
#Let's use Random Forest Classifier  
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier 
```

```python
#Let's define X and y values
y = df_general['churn']
X = df_general.drop(columns = 'churn')
```

```python
#Let's divide data to train and test
X_train , X_test , y_train, y_test = train_test_split(X,y,random_state = 42,test_size = 0.2)
```

```python
#Let's predict test values
rf = RandomForestClassifier().fit(X_train,y_train)
y_pred = rf.predict(X_test)
```

```python
#Let's evaluate the model
rf.score(X_test,y_test)
```
